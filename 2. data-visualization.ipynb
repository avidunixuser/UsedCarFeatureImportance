```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/data/gold/vehicles_gold.csv")
df['year']=df['year'].astype('int64')
df.drop(['id'],axis=1,inplace=True)
df.head()
```

```python
sns.pairplot(df2)
plt.title('Figure 1: Pair Plot')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/images/Insight-Pair-Plot',dpi=None)
plt.show()
```

```python
sns.distplot(df.price)
plt.tight_layout()
plt.title("Distribution of price")
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/images/Insight-Dist-Plot',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots()
fig.set_size_inches(5,5)

p1=sns.barplot(x='fuel',y='price',data=df,ax=axs)
axs.set_title('Figure 1')

plt.savefig('/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/images/Insight-Figure-1',dpi=None)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='fuel',y='price',hue='condition',data=df)
plt.title('Figure 2')
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/images/Insight-Figure-2',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(20,10)
sns.barplot(x='year',y='price',data=df,ax=axs[0])
sns.barplot(df.groupby('year').count()['price'].index,df.groupby('year').count()['price'].values,ax=axs[1])
axs[0].set_title('Figure 3')
axs[1].set_title('Figure 4')
axs[1].set_ylabel('Number of cars')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Application_II/practical_application_II_starter/images/Insight-Figure-3-4.jpg',dpi=600)
plt.show()
```
